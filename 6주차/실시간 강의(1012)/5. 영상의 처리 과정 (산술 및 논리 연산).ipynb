## 영상의 산술 연산
* 여러 영상들(2차원 행렬)의 **덧셈 및 뺄셈** 연산 ==> 새로운 결과 영상 생성
* 255 넘는 값들은 **포화연산(saturate())** 
* **cv2.add()** 이용 (덧셈 시, 전반적으로 밝게 포화되는 부분 많게 됨)
* 따라서 보통 **각 영상에 대해 a, b로 가중치** 부여!
    > **dst = saturate(a * src1 + b * src2 + r)**  (보통, **a + b = 1**)
* **cv2.addWeighted(src1, a, src2, b, r)** : 가중치 합 구하기 함수
    > **감마(r)** 인자를 통해 가중치의 합에 추가적인 덧셈까지 한번에 가능

```python
# cv2.addWeighted()

import cv2
import numpy as np

src1 = cv2.imread('airplane.bmp')
src2 = cv2.imread('field.bmp')

dst = np.empty(src1.shape, src1.dtype)

cv2.addWeighted(src1, 0.5, src2, 0.5, 0, dst) # 평균 밝기 유지됨

cv2.imshow('dst', dst)
cv2.imwrite('addWeighted.png', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

![img](addWeighted.png)

* 덧셈 연산과 달리, 뺄셈 순서에 따라 결과 달라짐
* if) 뺄셈 순서에 상관 없이 차이 연산을 구하고 싶다면? ==> **차영상(difference image)**
> **dst = |src1 - src2|** 
* ex) 틀린 그림 찾기
> **cv2.absdiff()** : 차영상 구하기

```python
# cv2.absdiff()

src1 = cv2.imread('lenna256.bmp', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('square.bmp', cv2.IMREAD_GRAYSCALE)

dst1 = np.empty(src1.shape, src1.dtype)
dst2 = np.empty(src1.shape, src1.dtype)

cv2.subtract(src1, src2, dst1)
cv2.absdiff(src1, src2, dst2) # 0에서 원본값 빼면 반전! (inverse)

cv2.imshow('subtract', dst1)
cv2.imwrite('subtract.png', dst1)
cv2.imshow('absdiff', dst2)
cv2.imwrite('absdiff.png', dst2)
cv2.waitKey()
cv2.destroyAllWindows()
```

## (실습) 영상의 산술 연산
* 두 이미지 파일을 이용하여 덧셈, 뺄셈, 평균, 차이 영상 생성 후 결과를 plot에 출력

```python
import matplotlib.pyplot as plt

src1 = cv2.imread('lenna256.bmp', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('square.bmp', cv2.IMREAD_GRAYSCALE)

dst1 = cv2.add(src1, src2)
dst2 = cv2.addWeighted(src1, 0.5, src2, 0.5, 0.0)
dst3 = cv2.subtract(src1, src2)
dst4 = cv2.absdiff(src1, src2)

plt.subplot(231), plt.axis('off'), plt.imshow(src1, 'gray'), plt.title('src1')
plt.subplot(232), plt.axis('off'), plt.imshow(src2, 'gray'), plt.title('src2')
plt.subplot(233), plt.axis('off'), plt.imshow(dst1, 'gray'), plt.title('add')
plt.subplot(234), plt.axis('off'), plt.imshow(dst2, 'gray'), plt.title('addWeighted')
plt.subplot(235), plt.axis('off'), plt.imshow(dst3, 'gray'), plt.title('subtract')
plt.subplot(236), plt.axis('off'), plt.imshow(dst4, 'gray'), plt.title('absdiff')
plt.show()
plt.close()
```

## 영상의 논리 연산
* **픽셀 값을 이진수로** 표현하여 **각 비트(bit) 단위 논리 연산** 수행
* cv2.bitwise_and() : AND (논리곱)
* cv2.bitwise_or() : OR (논리합)
* cv2.bitwise_xor() : XOR (배타적 논리합) ==> 두 개가 다르면 1, 같으면 0
* cv2.bitwise_not() : NOT (부정) ==> 반전시키기 (0 <-> 1)
> 각각의 함수들은 모두 **mask** 인자를 갖고 있음 ==> 0이 아닌 위치에서만 논리 연산 수행하도록 설정 가능!

* 그레이스케일 영상의 경우(uint8, 8bit), 8개의 비트에 모두 논리 연산
    * ex) 두 그레이스케일 값이 각각 110, 200 이면, 
    * 01101110(2) 와 11001000(2)의 논리 연산
    * 110 AND 200 = 01001000(2) = 72

**주의** : 각 픽셀(8bit) 마다 위의 논리 연산을 한 것이므로, 모두 1일때만 즉, 255일때만 흰색으로 나타나는 것! 
> 따라서 255와 255만이 논리곱했을 때 흰색으로 표현 됨!

```python
from matplotlib import pyplot as plt

src1 = cv2.imread('lenna256.bmp', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('square.bmp', cv2.IMREAD_GRAYSCALE)

print(np.max(src1)) # 최대값이 232 ==> 따라서 여기서 1이란 255 (흰색)!

plt.subplot(231), plt.axis('off'), plt.imshow(src1, 'gray'), plt.title('src1')
plt.subplot(232), plt.axis('off'), plt.imshow(src2, 'gray'), plt.title('src2')

dst1 = cv2.bitwise_and(src1, src2)
plt.subplot(233), plt.axis('off'), plt.imshow(dst1, 'gray'), plt.title('bitwise_and')
dst2 = cv2.bitwise_or(src1, src2)
plt.subplot(234), plt.axis('off'), plt.imshow(dst2, 'gray'), plt.title('bitwise_or')
dst3 = cv2.bitwise_xor(src1, src2)
plt.subplot(235), plt.axis('off'), plt.imshow(dst3, 'gray'), plt.title('bitwise_xor')
dst4 = cv2.bitwise_not(src1)
plt.subplot(236), plt.axis('off'), plt.imshow(dst4, 'gray'), plt.title('bitwise_not')

plt.show()

# bitwise_and : 255 (11111111)인 부분만 살아서 가운데 네모만 원본 값 그대로 출력
# bitwise_or : 둘 중에 하나라도 1이면 1이므로, 가운데 흰 부분은 다 11111111이라 흰색
# bitwise_xor : 두 값이 다르면 (1, 0) 1이므로, 가운데 네모에 0인 bit는 다 반전 (1<->0)
# bitwise_not : 걍 다 반전
```
